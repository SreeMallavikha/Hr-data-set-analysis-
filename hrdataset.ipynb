# Rework on Different Model

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
rfr=RandomForestRegressor(random_state=5)
rfr.fit(x_train,y_train)
```

```python
y1_pred=rfr.predict(x_test)
```

```python
#calculating mean absolute error,mean squared error ,root mean squared error 
print('Mean Absolute Error(MAE):',metrics.mean_absolute_error(y_test,y1_pred))
print('Mean Squared Error(MSE):',metrics.mean_squared_error(y_test,y1_pred))

print('Root Mean Squared Error (RMSE):',
      np.sqrt(metrics.mean_squared_error(y_test,y1_pred)))
mape=np.mean(np.mean(np.abs(y_test-y1_pred)/np.abs(y1_pred)))
mape1=round(mape*100,2)
accuracy1=round(100*(1-mape),2)
print('Mean Absolute Percentage Erroe (MAPE):',round(mape*100,2))
print('Accuracy:',round(100*(1-mape),2))
        
```
